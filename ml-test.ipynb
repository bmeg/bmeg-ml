```python
import sys
from bmeg.ml_schema_pb2 import Model
from google.protobuf import json_format
import pandas
from scipy import stats
from sklearn.linear_model import LinearRegression, LogisticRegression
import numpy as np
```

## Connect to BMEG

```python
O = ophion.Ophion("http://bmeg.io")
```

## List all cohorts

```python
cohorts = []
for a in O.query().has("type" ,"Cohort").execute():
    cohorts.append(a['gid'])
```

```python
cohorts
```

## Download RNA-Seq for cohort:TCGA-READ

```python
data = {}
for i in O.query().has("gid", "cohort:TCGA-READ").outgoing("hasSample").incoming("expressionFor").execute():
    if 'properties' in i and 'expressions' in i['properties']:
        data[i['gid']] = json.loads(i['properties']['expressions'])

expression_matrix = pandas.DataFrame(data).transpose().fillna(0.0)
```

```python
def exp_norm(expr):
    out_expr = expr.apply(
        lambda x: stats.expon.ppf((x.rank()-1) / len(x)), 1)
    return out_expr.fillna(0.0)
```

##Exponential 

```python
normalized_matrix = exp_norm(expression_matrix)
```

```python
model_data = {}
with open("ucsc-james.models.json") as handle:
    for line in handle:
        m = Model()
        json_format.Parse(line, m)

        feature_names = pandas.Index(m.structure.linear.coeff.keys()).intersection(normalized_matrix.columns)
        labels = pandas.Index(m.structure.linear.coeff.keys())
        s = LinearRegression()
        s.intercept_ = m.structure.linear.intercept
        s.coef_ = np.zeros(len(feature_names))
        for i, name in enumerate(feature_names):
            s.coef_[i] = m.structure.linear.coeff[name]
        model_row = dict(zip(normalized_matrix.index, s.predict(normalized_matrix.loc[:,feature_names])))
        model_data[m.id] = model_row
```
